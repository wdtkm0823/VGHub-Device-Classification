```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pixel3a_15s_soc50_home = pd.read_csv( '../VirtualGridHub/data_collection/eval_csv/dataset_15s/train/sink/GooglePixel3a#1/output_vghub_15s_data_3_pixel3a#1_soc50_home_train_1.csv')
ankerpowercore3_15s_soc50_home = pd.read_csv( '../VirtualGridHub/data_collection/eval_csv/dataset_15s/train/sink/AnkerPowerCore3#1/output_vghub_15s_data_3_ankerpowercore3#1_soc50_home_train_1.csv')
cheeropowermountain_15s_soc50_home = pd.read_csv( '../VirtualGridHub/data_collection/eval_csv/dataset_15s/train/sink/cheeroPowerMountain#1/output_vghub_15s_data_3_cheeropowermountain#1_soc50_home_train_1.csv')
cheeropowerplus5_15s_soc50_home = pd.read_csv( '../VirtualGridHub/data_collection/eval_csv/dataset_15s/train/sink/cheeroPowerPlus5#1/output_vghub_15s_data_3_cheeropowerplus5#1_soc50_home_train_1.csv')
ipadair4th_15s_soc50_home = pd.read_csv( '../VirtualGridHub/data_collection/eval_csv/dataset_15s/train/sink/iPadAir4th#1/output_vghub_15s_data_3_ipadair4th#1_soc50_home_train_1.csv')
jfpeace8_15s_soc50_home = pd.read_csv( '../VirtualGridHub/data_collection/eval_csv/dataset_15s/train/sink/JFPEACE8#1/output_vghub_15s_data_3_jfpeace8#1_soc50_home_train_1.csv')  

column_name = 'VBUS Current[A]'
pixel3a_15s_soc50_home_current = pixel3a_15s_soc50_home[column_name]
ankerpowercore3_15s_soc50_home_current = ankerpowercore3_15s_soc50_home[column_name]
cheeropowermountain_15s_soc50_home_current = cheeropowermountain_15s_soc50_home[column_name]
cheeropowerplus5_15s_soc50_home_current = cheeropowerplus5_15s_soc50_home[column_name]
ipadair4th_15s_soc50_home_current = ipadair4th_15s_soc50_home[column_name]
jfpeace8_15s_soc50_home_current = jfpeace8_15s_soc50_home[column_name]

column_name = 'VBUS Voltage[V]'
pixel3a_15s_soc50_home_voltage = pixel3a_15s_soc50_home[column_name]
ankerpowercore3_15s_soc50_home_voltage = ankerpowercore3_15s_soc50_home[column_name]
cheeropowermountain_15s_soc50_home_voltage = cheeropowermountain_15s_soc50_home[column_name]
cheeropowerplus5_15s_soc50_home_voltage = cheeropowerplus5_15s_soc50_home[column_name]
ipadair4th_15s_soc50_home_voltage = ipadair4th_15s_soc50_home[column_name]
jfpeace8_15s_soc50_home_voltage = jfpeace8_15s_soc50_home[column_name]

column_name = 'VBUS Wattage[W]'
pixel3a_15s_soc50_home_wattage = pixel3a_15s_soc50_home[column_name]
anckepoercore3_15s_soc50_home_wattage = ankerpowercore3_15s_soc50_home[column_name]
cheeropowermountain_15s_soc50_home_wattage = cheeropowermountain_15s_soc50_home[column_name]
cheeropowerplus5_15s_soc50_home_wattage = cheeropowerplus5_15s_soc50_home[column_name]
ipadair4th_15s_soc50_home_wattage = ipadair4th_15s_soc50_home[column_name]
jfpeace8_15s_soc50_home_wattage = jfpeace8_15s_soc50_home[column_name]

plt.rcParams['font.family'] = 'Arial'

fig, axs = plt.subplots(1, 3, figsize=(20, 5))

axs[0].plot(pixel3a_15s_soc50_home_current, label='Google Pixel 3a')
axs[0].plot(ankerpowercore3_15s_soc50_home_current, label='Anker PowerCore 3')
axs[0].plot(cheeropowermountain_15s_soc50_home_current, label='cheero Power Mountain')
axs[0].plot(cheeropowerplus5_15s_soc50_home_current, label='cheero Power Plus 5')
axs[0].plot(ipadair4th_15s_soc50_home_current, label='iPad Air 4th')
axs[0].plot(jfpeace8_15s_soc50_home_current, label='JFPEACE8')
axs[0].set_xlabel('Time (ms)', fontweight='bold')
axs[0].set_ylabel('VBUS Current (A)', fontweight='bold')
legend = axs[0].legend()
for text in legend.get_texts():
    text.set_fontweight('bold')
axs[0].grid(True)

axs[1].plot(pixel3a_15s_soc50_home_voltage, label='Google Pixel 3a')
axs[1].plot(ankerpowercore3_15s_soc50_home_voltage, label='Anker PowerCore 3')
axs[1].plot(cheeropowermountain_15s_soc50_home_voltage, label='cheero Power Mountain')
axs[1].plot(cheeropowerplus5_15s_soc50_home_voltage, label='cheero Power Plus 5')
axs[1].plot(ipadair4th_15s_soc50_home_voltage, label='iPad Air 4th')
axs[1].plot(jfpeace8_15s_soc50_home_voltage, label='JFPEACE8')
axs[1].set_xlabel('Time (ms)', fontweight='bold')
axs[1].set_ylabel('VBUS Voltage (V)', fontweight='bold')
legend = axs[1].legend()
for text in legend.get_texts():
    text.set_fontweight('bold')
axs[1].grid(True)

axs[2].plot(pixel3a_15s_soc50_home_wattage, label='Google Pixel 3a')
axs[2].plot(anckepoercore3_15s_soc50_home_wattage, label='Anker PowerCore 3')
axs[2].plot(cheeropowermountain_15s_soc50_home_wattage, label='cheero Power Mountain')
axs[2].plot(cheeropowerplus5_15s_soc50_home_wattage, label='cheero Power Plus 5')
axs[2].plot(ipadair4th_15s_soc50_home_wattage, label='iPad Air 4th')
axs[2].plot(jfpeace8_15s_soc50_home_wattage, label='JFPEACE8')
axs[2].set_xlabel('Time (ms)', fontweight='bold')
axs[2].set_ylabel('VBUS Wattage (W)', fontweight='bold')
legend = axs[2].legend()
for text in legend.get_texts():
    text.set_fontweight('bold')
axs[2].grid(True)

plt.tight_layout()
plt.savefig('./images/vghub_15s_deveice_soc50_home_cvw.png')
plt.show()
```
